```python
# 1. Import needed libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")
```

```python
# 2. Load the dataset
dataset = "data.csv"

df = pd.read_csv(dataset)
```

```python
# 3. Check df
df.head()
```

```python
# 4. Check basic statistics
print(f"Data shape: {df.shape}\n")

print("Data frame info:")
print(df.info())
print()

print("Age distribution:")
print(df["Age"].describe())
```

```python
# 5. Check Animal Distribution
plt.figure(figsize=(10, 4))

sns.countplot(x='Animal', data=df)

plt.title('Animal Counts', fontweight='bold', fontsize=16)

plt.xlabel('')
plt.xticks(rotation=20)

plt.ylabel('')

plt.grid(False)

for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.show()
```

```python
# 6. Age Distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

sns.histplot(df['Age'], bins=10, kde=True, ax=axes[0])
axes[0].set_title('Age Histogram', fontweight='bold', fontsize=16)
axes[0].set_xlabel('')
axes[0].set_ylabel('Frequency')
axes[0].grid(False)

sns.boxplot(x='Age', data=df, ax=axes[1])
axes[1].set_title('Age Boxplot', fontweight='bold', fontsize=16)
axes[1].set_xlabel('')
axes[1].set_ylabel('')
axes[1].grid(False)

plt.show()
```

```python
# 7. Age Distribution of each Animal
plt.figure(figsize=(14, 4))

sns.boxplot(x='Animal', y='Age', data=df)

plt.title('Age Distribution by Animal', fontweight='bold', fontsize=16)

plt.xlabel('')
plt.ylabel('Age')

plt.xticks(rotation=20)

plt.grid(False)

plt.show()
```
